```python
#  function to check whether the feature is gaussian distributed or not 
## we use Q-Q PLOT to check this 
import scipy.stats as stat
import pylab

def plot(dataset,feature):
    plt.subplot(1,2,1)
    dataset[feature].hist(bins =20)
    plt.subplot(1,2,2)
    stat.probplot(dataset[feature],plot=pylab)
    plt.show()
```
